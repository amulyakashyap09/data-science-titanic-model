```python
import pandas as pd
import numpy as np
import os
```

```python
# import data
# set the path of raw data

raw_data_path = os.path.join(os.path.pardir, 'data', 'raw')
train_file_path = os.path.join(raw_data_path, 'train.csv')
test_file_path = os.path.join(raw_data_path, 'test.csv')
```

```python
# read all the data with all default parameters

train_df = pd.read_csv(train_file_path, index_col='PassengerId')
test_df = pd.read_csv(test_file_path, index_col='PassengerId')
```

```python
test_df['Survived'] = -888
```

```python
df = pd.concat((train_df, test_df), axis=0)
```

```python
male_passengers=df.loc[df.Sex=="male",]
print("Number of male passengers {0}".format(len(male_passengers)))
```

```python
male_passengers=df.loc[(df.Sex=="male") & (df.Pclass == 1),]
print("Number of male passengers {0}".format(len(male_passengers)))
```

```python
df.describe()
```

```python
# mean
print('mean fare : {0}'.format(df.Fare.mean()))
# median
print('median fare : {0}'.format(df.Fare.median()))
```

```python
# measures
print('minimum fare : {0}'.format(df.Fare.min()))
print('maximum fare : {0}'.format(df.Fare.max()))

#range
print('range (max - min) fare : {0}'.format(df.Fare.max()-df.Fare.min()))

# Qaurtile
print('1st quartile fare : {0}'.format(df.Fare.quantile(.25)))
print('2nd quartile fare : {0}'.format(df.Fare.quantile(.50)))
print('3rd quartile fare : {0}'.format(df.Fare.quantile(.75)))

# variance
print('Fare Variance : {0}'.format(df.Fare.var()))
print('Fare Standard Deviation : {0}'.format(df.Fare.std()))
```
